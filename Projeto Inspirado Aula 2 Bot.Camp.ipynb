```python
# importar as bibliotecas e módulos

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
import yfinance as yf
```

```python
# requisição

driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))

url = "https://www.oceans14.com.br/acoes/busca-avancada"

driver.get(url)
```

```python
# achar elementos html
time.sleep(5)

botao_busca = driver.find_element("xpath", '''/html/body/form/div[4]/section[3]/div/div/div[1]/div[15]/button[2]''')

driver.execute_script("arguments[0].click();", botao_busca)

time.sleep(2)

botao_filtro = driver.find_element("xpath", '''/html/body/form/div[4]/section[3]/div/div/div[2]/div/div/div[2]/select''').click()
botao_todos = driver.find_element("xpath", '''/html/body/form/div[4]/section[3]/div/div/div[2]/div/div/div[2]/select/option[3]''').click()
```

```python
# copiar dados tabela de uma única vez #caso tenha outra pagina:  driver.find_element("xpath", '''xpath do número de páginas''')
# apagar para deixar apenas numérico variavel = variavel.text.replace("texto para repor", ""); int(variavel)

tabela = driver.find_element("xpath", 
                             '''/html/body/form/div[4]/section[3]/div/div/div[2]/div/div/div[1]/div[2]/div/div[2]/table''')

html_tabela = tabela.get_attribute("outerHTML")

# pandas serve para ler: excel, csv, html, o caralho todo

tabela_finalizada = pd.read_html(html_tabela, decimal = ',', thousands = '.')[0]

tabela_finali = tabela_finalizada.set_index("Papel")

tabela_final = tabela_finali[['P/L', 'P/VP', 'VPA','LPA','ROE', 'Div. Yield', 'Cresc. Receita (5 anos)']]

display(tabela_final)
```
